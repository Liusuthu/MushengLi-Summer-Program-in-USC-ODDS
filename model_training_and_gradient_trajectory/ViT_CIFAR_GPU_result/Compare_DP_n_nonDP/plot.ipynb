```python
import json
import matplotlib.pyplot as plt

#  JSON 
def read_json(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
    
    times = []
    gpu_state_1 = []
    gpu_state_2 = []
    
    for entry in data:
        times.append(entry['time'])
        gpu_state_1.append(entry['GPU_state'][0])
        gpu_state_2.append(entry['GPU_state'][1])
    
    return times, gpu_state_1, gpu_state_2

#  JSON 
times1, gpu_state_1_1, gpu_state_1_2 = read_json('non_DP.json')
times2, gpu_state_2_1, gpu_state_2_2 = read_json('DP.json')

#  Temperature 
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(times1, gpu_state_1_1, label='Data 1 - Temperature', marker='o')
plt.plot(times2, gpu_state_2_1, label='Data 2 - Temperature', marker='x')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Temperature over Time')
plt.legend()
plt.grid(True)

#  Memory Usage 
plt.subplot(2, 1, 2)
plt.plot(times1, gpu_state_1_2, label='Data 1 - Memory Usage', marker='o')
plt.plot(times2, gpu_state_2_2, label='Data 2 - Memory Usage', marker='x')
plt.xlabel('Time')
plt.ylabel('Memory Usage')
plt.title('Memory Usage over Time')
plt.legend()
plt.grid(True)

# 
plt.tight_layout()
plt.show()
```
