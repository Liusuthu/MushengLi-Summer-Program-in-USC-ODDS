```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt


loaded_true_gradients = np.load('nonDP_begin_true_gradients.npy', allow_pickle=True)
# 假设 gradient 是一个包含许多 numpy 数组的列表，并且这些数组的维度是相同的
gradient=[]
true_gradient=[]

for i in range(50):
    print('i=',i)
    true_grad=loaded_true_gradients[i]
    true_gradient.append(true_grad)
    loaded_per_sample_gradients = np.load(f'nonDP_begin_{i}_gradients.npy', allow_pickle=True)
    for j in range(20):
        per_sample_grad=loaded_per_sample_gradients[j]
        gradient.append(per_sample_grad)
```

```python
# 将列表转换为一个二维的 numpy 数组
gradient_matrix = np.vstack(gradient)
true_gradient_matrix = np.vstack(true_gradient)

# 使用 PCA 将数据降维到 2D
pca = PCA(n_components=2)
gradient_pca = pca.fit_transform(gradient_matrix)

# 获取投影矩阵
projection_matrix = pca.components_

# 对 true_gradient 进行投影
true_gradient_pca = pca.transform(true_gradient_matrix)

# 绘制结果
plt.scatter(gradient_pca[:, 0], gradient_pca[:, 1], alpha=0.5, label='Per-Sample Gradient')
plt.scatter(true_gradient_pca[:, 0], true_gradient_pca[:, 1],alpha=0.5, color='red', marker='x', s=100, label='True Gradient')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Gradients with True Gradient')
plt.legend()
plt.show()
```
